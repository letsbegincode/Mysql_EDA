```python
from bs4 import BeautifulSoup

# Read the HTML content from the file
with open("scrapped_Laptop_data.txt", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")
```

```python
blocks=soup.find_all('div',{'class':'cPHDOP'})
```

```python
len(blocks)
```

```python
print(blocks[0])
```

```python

# Initialize lists to store the extracted data
titles = []
ratings = []
rating_counts_and_reviews = []

curr_prices = []
prices = []
discounts = []

categories_list = []
processors_list = []
ram_capacities = []
ssd_capacities = []
screen_sizes = []
os_list = []

# Loop through the blocks to extract data
for block in blocks:
    # Extract title
    title_div = block.find('div', {'class': 'KzDlHZ'})
    titles.append(title_div.get_text(strip=True) if title_div else None)

    # Extract rating
    rating_span = block.find('div', {'class': 'XQDdHH'})
    ratings.append(rating_span.get_text(strip=True) if rating_span else None)

    # Extract rating count
    rating_count_div = block.find('span', {'class': 'Wphh3N'})
    rating_counts_and_reviews.append(rating_count_div.get_text(strip=True) if rating_count_div else None)

    # Extract current price
    curr_price_div = block.find('div', {'class': 'Nx9bqj _4b5DiR'})  # Adjust selector as needed
    curr_prices.append(curr_price_div.get_text(strip=True) if curr_price_div else None)

    # Extract original price
    price_div = block.find('div', {'class': 'yRaY8j'})  # Adjust selector as needed
    prices.append(price_div.get_text(strip=True) if price_div else None)

    # Extract discount
    discount_div = block.find('div', {'class': 'UkUFwK'})
    discounts.append(discount_div.get_text(strip=True) if discount_div else None)

    # Extract categories
    categories = block.find_all('li', {'class': 'J+igdf'})

    # Extract processor types
    processors_list.append(categories[0].get_text(strip=True) if len(categories) > 0 else None)

    # Extract RAM capacity
    ram_capacities.append(categories[1].get_text(strip=True) if len(categories) > 1 else None)

    # Extract operating system
    os_list.append(categories[2].get_text(strip=True) if len(categories) > 2 else None)

    # Extract SSD capacity
    ssd_capacities.append(categories[3].get_text(strip=True) if len(categories) > 3 else None)

    # Extract screen size
    screen_sizes.append(categories[4].get_text(strip=True) if len(categories) > 4 else None)
```

```python
import pandas as pd
# Create a DataFrame with the extracted data
df = pd.DataFrame({
    'Title': titles,
    'Rating': ratings,
    'Rating Count and Reviews': rating_counts_and_reviews,
    'Current Price': curr_prices,
    'Original Price': prices,
    'Discount': discounts,
    'Processor': processors_list,
    'RAM Capacity': ram_capacities,
    'Operating System': os_list,
    'SSD Capacity': ssd_capacities,
    'Screen Size': screen_sizes
})
```
